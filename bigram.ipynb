```python
g = torch.Generator().manual_seed(2147483647)

for i in range(10):
    out = []
    ix = 0
    
    while True:
        p = P[ix]
        # p = N[ix].float()
        # p = p/p.sum()
        ix = torch.multinomial(p, num_samples=1, replacement=True,generator=g)
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```

```python
log_likelyhood = 0.0
n=0
for w in words[:3]:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        prob = P[ix1,ix2]    #likelyhood is the product of these prob
        logprob = torch.log(prob) # -inf to 0  #log -> so that we can sum simply
        log_likelyhood += logprob
        n+=1
        print(f'{ch1}{ch2}: {prob:.4f} {logprob:.4f}')
print(f'{log_likelyhood=}')
nll = -log_likelihood
print(f'{nll=}') #loss
print(f'{nll/n}') #avg loss
```

```python
g = torch.Generator().manual_seed(2147483647)
W = torch.randn((27,27), generator = g, requires_grad=True)
```

```python
xenc = F.one_hot(xs, nums_classes=27).float()
logits = xenc @ W
counts = logits.exp()
probs = counts / counts.sum(1,keepdims=True)
probs
```
